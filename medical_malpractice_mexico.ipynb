```python
import requests ## request html for a page(s)
import pandas as pd ## pandas to work with data
from bs4 import BeautifulSoup
```

```python
url = "https://www.cndh.org.mx/palabras-clave/3175/negligencia-medica?page=0"
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text,"html.parser")
```

```python
# Get the text from the website 
soup.text
```

```python
# Organize the information. This is the container for all the elements in the pages
organized = soup.find("div", class_="col col-xs-12 col-sm-12 col-md-12 col-lg-12 view-grid-item")
organized
```

```python
# I'm cheching that my scrape for one single tittle in the first page out of the 4
# is working 
title_block = organized.find("div", class_ = "views-field views-field-title")
title_block 
```

```python
# I'm cleanning the scrape to just get the title text
only_tittle = title_block.find("span", class_="field-content")
only_tittle
```

```python
only_tittle.text
```

```python
# This is to get the link embeded in the title 
only_tittle.find ("a")["href"]
```

```python
tittle = only_tittle.text 
tittle_link = only_tittle.find ("a")["href"]
```

```python
##### NEXT STEPS NOW THAT I KNOW MY SINGLE SCRAPE WORKED 
```

```python
# TO GET THE RESTO OF THE TITTLES
# I'm using a for loop to get the rest of the titles in the 4 pages


rows = []
blocks = soup.find_all("div", class_="col col-xs-12 col-sm-12 col-md-12 col-lg-12 view-grid-item")

for block in blocks:
    title_block = block.find("div", class_ = "views-field views-field-title")
    only_tittle = title_block.find("span", class_="field-content")
    tittle = only_tittle.text 
    tittle_link = only_tittle.find ("a")["href"]
    rows.append([tittle, tittle_link])
print(rows)

    
```

```python
# I'm testing that I can get all the titles and links in one dataframe
for row in rows:
    print (row)
    
df = pd.DataFrame(rows, columns=["titulo", "titulo_link"])
df
```

```python
## For loop to get all the tittles in all the sites. Get a list with all my links

for page in range (3):
    url = f"https://www.cndh.org.mx/palabras-clave/3175/negligencia-medica?page={page}"
    response = requests.get(url)
    print(response.status_code)

```

```python
# New step. I'm puting a different for loops within the for loop. 
# I tested below that every new for loop and "if" statement for "title", "asunto", "programa"
# and "contenido" worked. Once I saw it worked, I moved the code to this big analysis 


rows = []
for page in range (3):
    url = f"https://www.cndh.org.mx/palabras-clave/3175/negligencia-medica?page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text,"html.parser")

    blocks = soup.find_all("div", class_="col col-xs-12 col-sm-12 col-md-12 col-lg-12 view-grid-item")
    
    for block in blocks:
        title_block = block.find("div", class_ = "views-field views-field-title")
        only_tittle = title_block.find("span", class_="field-content")
        tittle = only_tittle.text 
        tittle_link = only_tittle.find ("a")["href"]
        asunto_block = block.find("div", class_ = "views-field views-field-field-asunto")
        # This "if" "else" statement is to prevent the code from breaking
        # In case it finds dead ends. Some of the compleint didn't have 
        # "related content" or one of the other cathegories
        if asunto_block: 
            only_asunto = asunto_block.find("div", class_="field-content")
            asunto = only_asunto.text 
        else: 
            asunto = ""
            asunto_link = ""     
        programa_block = block.find("div", class_ = "views-field views-field-field-programa")
        if programa_block: 
            only_programa = programa_block.find("div", class_="field-content")
            programa = only_programa.text 
            programa_link = only_programa.find ("a")["href"]
        else: 
            programa = ""
            programa_link = ""
        contenido_block = block.find("div", class_ = "views-field views-field-field-comunicados")
        if contenido_block: 
            only_contenido = contenido_block.find("div", class_="field-content")
            contenido = only_contenido.text 
            contenido_link = only_contenido.find ("a")["href"]
        else: 
            contenido = ""
            contenido_link = ""
        # I'm appending all the information that I got in the for loops 
        #(title, tittle_link, etc) to the rows to get a data frame    
        rows.append([tittle, tittle_link, programa, programa_link, asunto, contenido, contenido_link])
print(rows)
```

```python
# I'm creating my data frame with my rows and columns. 
# The names within the quotes are the same variables I created in the for loops before
df = pd.DataFrame(rows, columns=["titulo", "titulo_link", "programa", "programa_link", "asunto", "contenido", "contenido_link"])
df
```

```python
# Time to export my data frama to a CSV file 
df.to_csv("denuncias.csv", encoding = "UTF-8", index = False)
```
